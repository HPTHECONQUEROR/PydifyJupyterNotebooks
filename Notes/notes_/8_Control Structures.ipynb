```python
#USING *args   - just a comment
def multi_elem(*args):  # defining a function ,"Hey compiler this is a function just note it down!"
    total = 0
    for num in args:
        total += num
    return total

var1 = multi_elem(1,2,3,4,5) # calling a function, "Hey compiler there is a function with the name multi_elem() just execute it with the given arguments 1,2,3,4,5"
var2 =multi_elem(12,43)
var1+var2
```

```python
def combine(*args):
    result = ""
    for string in args:
        result += string
    return result

concatenated = combine("Hello, ", "man!", " How", " are", " you?","Hello","Bro")
print( concatenated)
```
